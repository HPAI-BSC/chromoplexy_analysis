# Example of use


```python
from graph_pattern_extraction import generate_all_patient_graphs, call_gspan_from_library
from graph_pattern_extraction import save_graphs_to_pdf, save_graphs_to_png
```

First we define the wanted distances used to generate the graphs.

```python
distances = [100, 500, 1000]
```

Then we generate the subgraph files fore every distance. 

It reads from *data/allpatients/* and saves them into *data/allfiles_gspan/gspan_subgraphs_w100.txt* 

```python
for max_distance in distances:
    generate_all_patient_graphs(max_distance)
```

Extract the common subgraphs of the wanted distance using gspan and save it to a file.

It reads from *data/allfiles_gspan/gspan_subgraphs_w100.txt* and saves it to *data/subgraphs/gspan_subgraphs_w100_frequent_subgraphs.txt*

```python
for d in distances:
    f = '../data/allfiles_gspan/gspan_subgraphs_w' + str(d) + '.txt'
    call_gspan_from_library(f)
```

Plot the common subgraphs into pdf and png and saves them in *data/plots/gspan_subgraphs_w100_frequent_subgraphs.pdf*


```python
for d in distances:
    f = '../data/subgraphs/gspan_subgraphs_w' + str(d) + '_frequent_subgraphs.txt'
    save_graphs_to_pdf(f)
    save_graphs_to_png(f)
```
